# 1. Import Libraries

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

# 2. Loading CSV

```python
df = pd.read_csv('Lesson_4_student_grades_dataset.csv')
df
```

# 3. Visualize the Data

```python
plt.scatter(df['Previous_Exam_Scores'],df['Grades'])
plt.xlabel('Previous_Exam_Scores')
plt.ylabel('Grades')
plt.title('Scatter Plot of Data')
plt.show()
```

# 4. Preparing Data for Modeling

```python
Previous_Exam_Scores = df[['Previous_Exam_Scores']]
Grades = df[['Grades']]
```

# 5. Fitting the Linear Regression Model

```python
model = LinearRegression()

model.fit(Previous_Exam_Scores,Grades)
```

# 6. Getting Model Coefficients

```python
slope = model.coef_[0]

intercept = model.intercept_

print('Slope:',slope)
print('Intercept:',intercept)
```

# 7. Visualizing the Linear Regression Line

```python
plt.scatter(Previous_Exam_Scores,Grades)
plt.plot(Previous_Exam_Scores,model.predict(Previous_Exam_Scores), color='yellow')
plt.xlabel('Previous_Exam_Scores')
plt.ylabel('Grades')
plt.title('Scatter Plot of Data')
plt.show()
```

# 8. Making Prediction

```python
new_X = [[88]]

predicted_Y = model.predict(new_X)

print('Value of X:',new_X)
print('Predicted Value of Y:',predicted_Y)
```
