1. Data Loading

```python
import pandas as pd


df = pd.read_csv('hardwareStore.csv')
df
```

2. Initial Data Analysis

```python
mean_value = df.mean()
median_value = df.median()
std_deviation = df.std()

print("Mean:")
print(mean_value)

print("Median:")
print(median_value)

print("Standard Deviation:")
print(std_deviation)
```

```python
unique = df['CATEGORY_NAME'].unique()
uniquee = df['PRODUCT_NAME'].unique()
uniqueeee = df['ADDRESS'].unique()

print("Unique categories and their counts:")
print(unique)
print(uniquee)
print(uniqueeee)

```

3. Category Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sne

summary_stats = df[['CATEGORY_NAME','PRODUCT_NAME']].describe()

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sne.histplot(df['CATEGORY_NAME'],bins=15,kde=True)
plt.title('Distribution of Products Across Categories')
```

```python
correlation_matrix = df[['CATEGORY_ID','PRODUCT_ID']].corr()

plt.figure(figsize=(8,6))
sne.heatmap(correlation_matrix,annot=True,cmap='coolwarm')
plt.title('Correlation Heatmap')
```

4. Cost and Price Analysis

```python
import matplotlib.pyplot as plt
import seaborn as sne

summary_stats = df[['STANDARD_COST','LIST_PRICE']].describe()

plt.figure(figsize=(12,5))

plt.subplot(1,2,1)
sne.histplot(df['STANDARD_COST'],bins=15,kde=True)
plt.title('AVERAGE STANDARD COST AND AVERAGE LIST PRICE')
```

```python
max_value = df['STANDARD_COST'].max()
max_valuee = df['LIST_PRICE'].max()

print(max_value)
print(max_valuee)
```

5. Location Analysis

```python
summary_stats = df[['STATE','PRODUCT_ID']].describe()

plt.figure(figsize=(30,5))

plt.subplot(1,2,1)
sne.histplot(df['STATE'],bins=15,kde=True)
plt.title('CITY WITH THE HIGHEST QUANTITY OF PRODUCTS')
```

6. INSIGHTS AND REPORTING

```python
report = f"""
***Explanatory Data Analysis Report***
***Summary Statistics***
{mean_value}
{median_value}
{std_deviation}
{unique}
{uniquee}
{uniqueee}
{max_value}
{max_valuee}
"""

print(report)
```
